```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("nba_2016_2017_100.csv", error_bad_lines=False)
data2 = pd.read_csv("nba_2017_nba_players_with_salary.csv", error_bad_lines=False)

#print data
```

```python
df1 = data[['PLAYER_NAME', 'TWITTER_FOLLOWER_COUNT_MILLIONS']]
df5 = data
df4 = data2
```

```python
# Code for making Top 10 hiatogram 

df1 = df1.sort_values('TWITTER_FOLLOWER_COUNT_MILLIONS', ascending=False)
top10 = df1[:10]
twitter10 = [(top10.iloc[i][0], top10.iloc[i][1]) for i in range(10)]
import matplotlib.pyplot as plt

c = [d[0] for d in twitter10]
c1 = [d[1] for d in twitter10]

plt.bar(c[:10], c1[:10])
plt.xlabel("Players")
plt.ylabel("Followers (Millions)")
plt.title("Top 10 NBA Players by Twitter Followers 2016-17")
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Code for creating scatterplots
import seaborn as sns

g = sns.jointplot(x="SALARY_MILLIONS", y="TWITTER_FOLLOWER_COUNT_MILLIONS", data=df5, kind='reg', color = 'b')
regline = g.ax_joint.get_lines()[0]
regline.set_color('red')
regline.set_zorder('5')
import numpy as np
```

```python
# Code for Linear Regression Tool

defense = smf.ols('TWITTER_FOLLOWER_COUNT_MILLIONS ~ SALARY_MILLIONS', data=df5).fit()
print(defense.summary())
```
